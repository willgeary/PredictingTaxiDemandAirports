```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../../clean_data/passenger_waittimes.csv')
```

```python
df.head()
```

```python
def calc_PAX_through(row):
    if row[6] == 0:
        return 0
    elif row[5] == 0:
        return row[6]
    else:
        return 5.0*row[6]/row[5]
```

```python
df['PAX_through'] = df.apply(calc_PAX_through, axis=1)
```
