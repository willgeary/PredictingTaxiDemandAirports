```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../../clean_data/passenger_waittimes.csv')
```

```python
df.head(400)
```

```python
def calc_PAX_through(row):
    if row[6] == 0:
        return 0
    elif row[5] == 0:
        return row[6]
    else:
        return 5.0*row[6]/row[5]
```

```python
df['PAX_through'] = df.apply(calc_PAX_through, axis=1)
```

```python
df.dtypes
```

```python
df.head()
```

```python
df[df['Max Total Time'] == 0]
```

```python
df['passengers_waiting'] = df['Max # of PAX in Line'] - df['PAX_through']
```

```python
df.head(400)
```

```python
df = df.set_index('dt')
```

```python
df.head()
```

```python
df.index = pd.DatetimeIndex(df.index)
```

```python
df.head()
```

```python
# df.dt = pd.to_datetime(df.dt)
```

```python
df[df['passengers_waiting'] < 0] = 0
```

```python
df
```

```python
df = df.resample('1H').last()
```

```python
df
```

```python
out = df.groupby([df.index.hour])[["passengers_waiting"]].mean()
```

```python
out
```

```python
out.to_csv('../../clean_data/biases_per_hour.csv')
```
